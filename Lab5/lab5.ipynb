1. Korzystając z równania parametrycznego narysuj sferę jednostkową w 3D

v =

    [ cos(s) sin(t) ]

    [ sin(s) sin(t) ]

    [ cos(t) ]

s ∈ [0, 2π], t ∈ [0, π]

```python
plt.figure()
plot_axis = plt.axes(projection='3d')
u,v = np.mgrid[0:np.pi:40j, 0:2*np.pi:30j]
x = np.cos(u) * np.sin(v)
y = np.sin(u) * np.sin(v)
z = np.cos(v)
plot_axis.scatter3D(x, y, z)
```

```python
n = 1000
plt.figure()
plot_axis = plt.axes(projection='3d')
s = np.random.uniform(0,np.pi,size = n)
t = np.random.uniform(0,2*np.pi,size = n)
x = np.cos(s) * np.sin(t)
y = np.sin(s) * np.sin(t)
z = np.cos(t)
plot_axis.scatter3D(x, y, z)
```

```python
plt.figure()
plot_axis = plt.axes(projection='3d')
v = np.array([x,y,z]).T
A = np.random.uniform(0,10,size = [3,3])
v = v @ A
v = v.T
plot_axis.scatter3D(v[0], v[1], v[2])
```

```python
ascent = scipy.misc.ascent()
# arr = np.asarray(ascent)
plt.gray()
plt.imshow(ascent)
```

```python
U,s,Vh = linalg.svd(ascent)
```

```python
# s
k = 30
s_len = len(s)
B = np.asarray(scipy.linalg.diagsvd(s, s_len, s_len))
# print(len(B[0]))
# print(len(B[0]))
for i in range(k,s_len):
     B[i][i] = 0

# for i
# in range(k,len(s)):
#      s[i] = 0
new_A = np.asarray(U @ B @ Vh)
# new_A
# # # # new_A
plt.imshow(new_A)
```

```python
def generate():
     nn = 100
     tab = []
     for k in range(nn):
          s_len = len(s)
          B = np.asarray(scipy.linalg.diagsvd(s, s_len, s_len))
          for i in range(k,s_len):
               B[i][i] = 0
          new_A = np.asarray(U @ B @ Vh)
          tab.append(np.sum(abs(new_A-ascent)))
     return tab
```
